```python
import cv2 
import os 

```

```python
# List of video paths
video_paths = [
    "C:/Users/jan/Downloads/happy1.mp4",
    "C:/Users/jan/Downloads/happy2.mp4"
]

try: 
    # creating a folder named data 
    if not os.path.exists('happydata'): 
        os.makedirs('happydata') 

# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of data') 

# Iterate over each video
for video_path in video_paths:
    # Read the video from specified path 
    cam = cv2.VideoCapture(video_path) 

    # frame 
    currentframe = 0

    while(True): 
        # reading from frame 
        ret, frame = cam.read() 

        if ret: 
            # if video is still left continue creating images 
            name = './happydata/' + os.path.basename(video_path) + '_frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name) 

            # writing the extracted images 
            cv2.imwrite(name, frame) 

            # increasing counter so that it will 
            # show how many frames are created 
            currentframe += 1
        else: 
            break

    # Release the current video capture once done 
    cam.release() 

cv2.destroyAllWindows()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Function to load and preprocess images
def load_and_preprocess(folder_path, label):
    images = []
    labels = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"):  
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)
            img = cv2.resize(img, (28, 28))  
            img = img / 255.0  # Normalize pixel values to the range [0, 1]
            images.append(img)
            labels.append(label)
    return images, labels

# Load and preprocess data
happy_images, happy_labels = load_and_preprocess("happydata", 0)
sad_images, sad_labels = load_and_preprocess("saddata", 1)
angry_images, angry_labels = load_and_preprocess("angrydata", 2)

# Combine data from all classes
all_images = np.concatenate((happy_images, sad_images, angry_images), axis=0)
all_labels = np.concatenate((happy_labels, sad_labels, angry_labels), axis=0)

# Convert lists to numpy arrays
all_images = np.array(all_images)
all_labels = np.array(all_labels)
```

```python
# Create an ImageDataGenerator for data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(all_images, all_labels, test_size=0.2, random_state=42)
```

```python
# Designing the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(3, activation='softmax')  # Changed to 3 output units for multiclass classification
])
```

```python
# Compiling the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Changed loss function for multiclass classification
              metrics=['accuracy'])
```

```python
# Training the model
model.fit(datagen.flow(X_train, y_train, batch_size=10), epochs=10, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
accuracy = model.evaluate(X_test, y_test)[1]
print(f"Test Accuracy: {accuracy}")
```

```python
# Save the model
model.save('emotion_detection_model.h5')
```

```python
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import numpy as np
import cv2
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model("emotion_detection_model.h5")

# Define class labels
class_labels = ['Happy', 'Sad', 'Angry']

# Function to preprocess the uploaded image
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (28, 28))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Function to predict emotion from the uploaded image
def predict_emotion(image_path):
    img = preprocess_image(image_path)
    prediction = model.predict(img)
    predicted_class = np.argmax(prediction)
    emotion = class_labels[predicted_class]
    return emotion

# Function to handle image upload
def upload_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        emotion = predict_emotion(file_path)
        display_image(file_path, emotion)

# Function to display the uploaded image and predicted emotion
def display_image(image_path, emotion):
    img = Image.open(image_path)
    img = img.resize((250, 250), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    panel = tk.Label(root, image=img)
    panel.image = img
    panel.pack()
    
    emotion_label.config(text="Predicted Emotion: " + emotion)

# Create the main application window
root = tk.Tk()
root.title("Emotion Recognition")

# Set the size of the window
root.geometry("500x400")

# Create a button to upload an image
upload_button = tk.Button(root, text="Upload Image", command=upload_image)
upload_button.pack(pady=20)

# Create a label to display predicted emotion
emotion_label = tk.Label(root, text="")
emotion_label.pack()

# Run the application
root.mainloop()
```
